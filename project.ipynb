```python
question_bank = {
    'Easy': [
        {
            'question': 'What is the capital of France?',
            'options': ['Berlin', 'Madrid', 'Paris', 'Rome'],
            'answer': 'Paris',
            'marks': 10,
            'negative_marks': -5
        },
        {
            'question': 'Which planet is known as the Red Planet?',
            'options': ['Earth', 'Mars', 'Jupiter', 'Venus'],
            'answer': 'Mars',
            'marks': 10,
            'negative_marks': -5
        },
        {
            'question': 'What is 2 + 2?',
            'options': ['3', '4', '5', '6'],
            'answer': '4',
            'marks': 10,
            'negative_marks': -5
        }
    ],
    'Medium': [
        {
            'question': 'What is the chemical symbol for gold?',
            'options': ['Ag', 'Au', 'Fe', 'Pb'],
            'answer': 'Au',
            'marks': 20,
            'negative_marks': -10
        },
        {
            'question': 'Who wrote \"To Kill a Mockingbird\"?',
            'options': ['Harper Lee', 'Mark Twain', 'J.K. Rowling', 'Ernest Hemingway'],
            'answer': 'Harper Lee',
            'marks': 20,
            'negative_marks': -10
        },
        {
            'question': 'What is the largest ocean on Earth?',
            'options': ['Atlantic', 'Indian', 'Arctic', 'Pacific'],
            'answer': 'Pacific',
            'marks': 20,
            'negative_marks': -10
        }
    ],
    'Hard': [
        {
            'question': 'Which year did the Titanic sink?',
            'options': ['1905', '1912', '1918', '1923'],
            'answer': '1912',
            'marks': 30,
            'negative_marks': -15
        },
        {
            'question': 'Who developed the theory of relativity?',
            'options': ['Isaac Newton', 'Marie Curie', 'Albert Einstein', 'Galileo Galilei'],
            'answer': 'Albert Einstein',
            'marks': 30,
            'negative_marks': -15
        },
        {
            'question': 'What is the longest river in the world?',
            'options': ['Amazon', 'Nile', 'Yangtze', 'Mississippi'],
            'answer': 'Nile',
            'marks': 30,
            'negative_marks': -15
        }
    ]
}

def pseudo_shuffle(lst):
    """
    Shuffles a list using a pseudo-random mechanism without 'random' module.
    """
    n = len(lst)
    if n <= 1:
        return list(lst)

    shuffled_list = list(lst)

    seed = 0
    for item in shuffled_list:
        seed += hash(str(item)) % 1000

    for i in range(n):
        target_index = (seed + i * 7) % n
        shuffled_list[i], shuffled_list[target_index] = shuffled_list[target_index], shuffled_list[i]

    return shuffled_list

def run_quiz(questions_to_ask):
    """
    Administers a quiz, collects answers, and calculates score with negative marking.
    """
    total_score = 0
    quiz_results = {
        'total_score': 0,
        'question_scores': [],
        'participant_answers': []
    }

    print("\n--- QUIZ START ---")
    print("Please enter the letter corresponding to your chosen answer (e.g., 'a', 'b', 'c', 'd').")



    option_map = {'a': 0, 'b': 1, 'c': 2, 'd': 3}

    for i, question_data in enumerate(questions_to_ask):
        print(f"\nQuestion {i + 1}: {question_data['question']}")
        for idx, option in enumerate(question_data['options']):
            option_letter = chr(ord('a') + idx)
            print(f"  {option_letter}) {option}")

        user_answer_raw = input("Your answer: ").strip().lower()

        chosen_answer_text = "Invalid Input"
        is_valid_input = False

        if user_answer_raw in option_map:
            chosen_option_index = option_map[user_answer_raw]
            if 0 <= chosen_option_index < len(question_data['options']):
                chosen_answer_text = question_data['options'][chosen_option_index]
                is_valid_input = True

        current_question_score = 0
        is_correct = False

        if is_valid_input and chosen_answer_text == question_data['answer']:
            current_question_score = question_data['marks']
            total_score += question_data['marks']
            is_correct = True
            print(f"Correct! You get {question_data['marks']} points.")
        elif is_valid_input:
            current_question_score = question_data['negative_marks']
            total_score += question_data['negative_marks']
            print(f"Incorrect. The correct answer was '{question_data['answer']}'. You lose {abs(question_data['negative_marks'])} points.")
        else:
            print("Invalid input. No points awarded or deducted for this question.")

        quiz_results['question_scores'].append({
            'question': question_data['question'],
            'your_answer': chosen_answer_text,
            'correct_answer': question_data['answer'],
            'marks_awarded': current_question_score,
            'is_correct': is_correct
        })
        quiz_results['participant_answers'].append(chosen_answer_text)

    quiz_results['total_score'] = total_score
    print("\n--- QUIZ END ---")
    print(f"Final Score: {total_score}")

    return quiz_results


difficulty_levels_to_test = ['Easy', 'Medium', 'Hard']

for difficulty in difficulty_levels_to_test:
    print(f"\n\n--- Running Quiz for Difficulty: {difficulty.upper()} ---")

    questions_for_quiz = question_bank[difficulty]
    shuffled_quiz_questions = pseudo_shuffle(questions_for_quiz)
    final_results = run_quiz(shuffled_quiz_questions)

    print(f"\nDetailed Results for {difficulty.upper()} Quiz:")
    for res in final_results['question_scores']:
        print(f"- Question: {res['question']}")
        print(f"  Your Answer: '{res['your_answer']}' ({'Correct' if res['is_correct'] else 'Incorrect'}) ")
        print(f"  Correct Answer: '{res['correct_answer']}'")
        print(f"  Marks Awarded: {res['marks_awarded']}")
        print("--------------------------------------------------")
```
